## 【母平均の区間推定】

・一部分の標本から全体の集団(母集団)の平均値や分散を、「確率」を利用して予想する事を、「推定」という

・「推定」には「点推定」「区間推定」が存在する。

・前者は１つの値を「推定値」にし、後者はある区間に求める母数(平均値、分散)がはいっているというように、ある「幅」をもって推定する方法で最も利用されている。

10人の標本から全国平均を区間推定
    母集団の平均を標本の値から適当な幅をもたせて推測する、というのが「母平均の区間推定」
    区間推定の場合、推定の「精度」を$$100(1-α)%$$のようにパーセントで表し、これを「信頼度」という。
   
母集団が、正規分布で近似されるとき、母平均を区間推定する。
正規母集団$N(μ, σ^2)$の母数の平均μと分散$σ^2$があるが、この分散が既知の場合と未知の場合では区間推定方法が異なる
どちらも母集団が正規分布に従っているという情報が必要。

## 【1. 母分散が既知の場合の区間推定方法】

--------

### 【1.1 : 母平均の信頼度限界(母分散既知)】
$$σ^2=\frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2$$

$$CL = \bar{X}\pm k\sqrt{\frac{σ^2}n}$$

$$\begin{equation}k=CL\left \{\begin{array}{l}  90\%=1.65 \\ 95\%=1.96\end{array}\right.\end{equation}$$
標本平均$\bar{X}$の分布は$N(μ, σ^2/\sqrt{n})$なので、$\bar{X}$を標準化すると
$$Z=\frac{\bar{X}-μ}{\sqrt{σ^2/n}}$$の分布は$N(0, 1)$になる。右袖確立2.5%点は、1.96なので、Zが含まれる確率が95%になる原点対称区間は
$$P(-1.96 \le Z \le 1.96) = 0.95$$
この区間を$μ$に関する95%区間に変形
$$P=\begin{pmatrix} \bar{X}-1.96\frac{σ}{\sqrt{n}}\le μ \le \bar{X}+1.96\frac{σ}{\sqrt{n}}\end{pmatrix}=0.95$$
このμにかんする区間を信頼区間と言い、以下
$$\begin{eqnarray}
\left[
\begin{array}{ccc}
\bar{X}-1.96\frac{σ}{\sqrt{n}}\le,　\le \bar{X}+1.96\frac{σ}{\sqrt{n}}
\end{array}
\right]
\end{eqnarray}$$
また、0.95(95%)を信頼係数と呼ぶ、これは100回区間推定を行うとその内95回は真の母数がこの区間にはいるという意味を持つ
一般的な信頼係数$100(1-α)\%$では、標準正規分布の右袖確立を$z_{α/2}$とすれば
$$\begin{eqnarray}
\left[
\begin{array}{ccc}
\bar{X}-z_{α/2}\frac{σ}{\sqrt{n}}\le,　\le \bar{X}+z_{α/2}\frac{σ}{\sqrt{n}}
\end{array}
\right]
\end{eqnarray}$$

----

### 【1.2 : 母平均の信頼限界度(母分散が未知)】
$$\bar{X}=\frac{\sum_{i=1}^{n}}{n}$$

$$S^2=\frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2$$
母分散$σ^2$を$S^2$に代えると、\bar{X}の標準統計量
$$T=\frac{\bar{X}-μ}{\sqrt{S^2/n}}$$
これは、自由度が(n-1)の　$t分布$　を持ち、信頼区間の構成には $t(n-1)$分布表を使う。
信頼係数95%の信頼区間作成には、t(n-1)の右2.5%が必要であり、これは$t_{0.025}(n-1)$であり、内側確率が0.95、つまり
$$P(-t_{0.025}(n - 1) \le T \le t_{0.025}(n - 1))= 0.95$$
であるので、Tの定義をもとに、μに関する不等式を導けば、
$$P=\begin{pmatrix} \bar{X}-t_{0.025}(n-1)\frac{S}{\sqrt{n}}\le μ \le \bar{X}+t_{0.025}(n-1)\frac{S}{\sqrt{n}}\end{pmatrix}=0.95$$
母平均μの95%信頼区間は
$$\begin{eqnarray}
\left[
\begin{array}{ccc}
\bar{X}-t_{0.025}(n-1)\frac{S}{\sqrt{n}}\le,　\le \bar{X}+t_{0.025}(n-1)\frac{S}{\sqrt{n}}
\end{array}
\right]
\end{eqnarray}$$
信頼係数が$100(1-α)\%$の場合、μの100(1-α)%信頼区間は、t(n-1)分布の右確率α/2、$t_{α/2}(n-1)$で、
$$\begin{eqnarray}
\left[
\begin{array}{ccc}
\bar{X}-t_{α/2}(n-1)\frac{S}{\sqrt{n}}\le,　\le \bar{X}+t_{α/2}(n-1)\frac{S}{\sqrt{n}}
\end{array}
\right]
\end{eqnarray}$$

$$\begin{equation}k=\left \{\begin{array}{l}  90\%=1.65 \\ 95\%=1.96\end{array}\right.\end{equation}$$

$$母平均の信頼限界=80 \pm 1.96\sqrt{\frac{5^2}{10}}$$
$$80 + 3.1 = 83.1$$
$$80 - 3.1 = 76.9$$

## 母比率の区間推定

母比率の区間推定を求める場合、標本の数が十分大きい時には近似的に標準正規分布に従う事が知られている。

1. 母平均の信頼限界(標本数 $\geqq 30$)

$$\bar{R}=標本比率$$

$$\bar{R}\pm k \sqrt{\frac{\bar{R}(1-\bar{R})}{\bar{n}}}$$

$$\begin{equation}k=\left \{\begin{array}{l}  90\%=1.65 \\ 95\%=1.96\end{array}\right.\end{equation}$$

(例１)：ある地域の有権者160人を調査。40人が「Aを支持」。この地域のA政党の支持率を95%信頼区間で推定する。

標本比率 = $\frac{40}{160}=0.25$

標本数 = 160

公式により母比率の信頼限界
$$0.25 \pm 1.96\sqrt{\frac{0.25(1-0.25)}{160}}$$
$$0.25 + 0.067 = 0.317$$
$$0.25 - 0.067 = 0.183$$
つまり、この地域のA政党の支持率は信頼度95%で$18.3\% ~ 31.8\%$の間にあるといえるので、票のゆくえは定かではない。

(例２)：ビデオリサーチ、300世帯の標本のうち、120世帯が特定の番組にチャンネルを合わせたとする。すると視聴率は
$$\frac{120}{300} = 標本比率 0.40$$

信頼区間(90%)
$$0.40 \pm 1.65 * \sqrt{\frac{0.40(1-0.40)}{300}} = 0.047$$

$$0.40\pm0.047 = 0.353 \backsim 0.447$$

つまり信頼区間90％では35.2%~44.6%の間にあるといえる。

信頼区間(95%)
$$0.40 \pm 1.96 * \sqrt{\frac{0.40(1-0.40)}{300}} = 0.055$$

$$0.40\pm0.055 = 0.345 \backsim 0.455$$

信頼区間95%では34.4%~45.4%の間にある

## 正規母集団の分散の区間推定

$N(μ, σ^2)$からサイズnの無作為標本　${x_1, X_2, ..., X_n}$　を得る、母分散$σ^2$について、信頼係数95%の区間推定を行う。

$σ^2$の推定量は
$$S^2 = \frac{1}{n-1}\sum_{n=1}^{n}(x_i - \bar{X})$$

$$U = \frac{\sum_{i=1}^{m}(X_i - \bar{X})^2}{σ^2}=\frac{(n - 1)S^2}{σ^2}$$

これは、自由度(n-1)のカイ二条分布に従う。信頼敬意数95% = α = 0.05
右97.5％点と2.5%点を、各々$U_{1-α/2}$, $U_{α/2}$

$$P(U_{1-α/2} < U < U_{α/2}) = 0.95$$

Uの95%区間が構成出来る。したがって上記式を代入
$$P=\begin{pmatrix}U_{1-α/2} < \frac{(n-1)S^2}{σ^2} < U_{α/2} \end{pmatrix}$$

$$ = P \begin{pmatrix}\frac{(n-1)S^2}{U_{α/2}} < σ^2 < \frac{(n-1)S^2}{U_{1-α/2}} \end{pmatrix}$$

$σ^2$95%信頼区間
$$
\begin{eqnarray}
\left[
\begin{array}{ccc}
\frac{(n-1)S^2}{U_{α/2}},　\frac{(n-1)S^2}{U_{1-α/2}}
\end{array}
\right]
\end{eqnarray}$$


(例１) : TOPIXの日次収益率の分散について、リターン分析が正規分布であるとして95%信頼区間を推定する

平均の区間推定から行う　$\bar{X}= 0.1154$ 、 $S=\sqrt{1.4675} = 1.2114$ 

分散が未知なので、標本分散$S^2$を使う。しかし、観測個数が　$\geqq 30$　なのでｔ分布ではなく

$$T=\frac{\bar{X}-μ}{\sqrt{S^2/n}}$$

は、中心極限定理により、標準正規分布で近似可能。

信頼係数95%なので、右2.5％の1.96を用いると、信頼区間は

$$
\begin{eqnarray}
\left[
\begin{array}{ccc}
0.1154 - 1.96 * \frac{1.2114}{10},　0.1154 + 1.96 * \frac{1.2114}{10}
\end{array}
\right]
\end{eqnarray} = [-0.122, 0.353]$$

分散の区間推定は、$S^2=1.4675$ を使うので、カイ二乗の分位点は、自由度が100-1=99と大きいので、表から自由度100の値の近似値を使う。
つまり自由度99のカイ二乗分布の分位点を正確に求めた方がよい

この場合　右97.5% = 74.22193, 右2.5% = 129.5612

$$
\begin{eqnarray}
\left[
\begin{array}{ccc}
\frac{(n-1)S^2}{U_{α/2}},　\frac{(n-1)S^2}{U_{1-α/2}}
\end{array}
\right]
 = 
\left[
\begin{array}{ccc}
\frac{99 * 1.4675}{129.5612},　\frac{99 * 1.4675}{74.22193}
\end{array}
\right]
\end{eqnarray} = [1.1213, 1.9574]$$

# 【母平均推定の標本数】


母集団から標本を抜き出して区間推定する時、問題になるのが、**「標本の数をどの程度にすればいいのか」**

例えば、街頭調査の場合など、何件のデータを取れば十分なのか、わかりにくい、コストにもはねかえる問題なだけに「コストは安く」
データの信頼度は確実にしたい。

標本の数は多ければ多いほどよいが、経費や時間などの点から標本の数は制限される。
もちろん、少なすぎても正確さに欠けた結果しか得られない。

区間推計の誤差に注目して、標本の数を決める。

母平均推定の標本数(簡易)
$$標本数 =\begin{pmatrix} \frac{k}{誤差の比率}*　標準偏差 　　　\end{pmatrix}^2$$

$$\begin{equation}k=\left \{\begin{array}{l}  90\%=1.65 \\ 95\%=1.96　\\ 99\%=2.58　\end{array}\right.\end{equation}$$


(例)紙の長さの誤差の標準偏差は、これまでの経験から3メートル。紙の平均の長さの99%信頼区間を求める時、誤差の比率を0.9にするには、標本の数をどの程度にすればよいか

$標準偏差=3$, $誤差の比率=0.9$, $k=2.58$

$$標本数=\begin{pmatrix} \frac{2.58}{0.9}*3 \end{pmatrix}^2 = 73.96$$

したがって約74個の紙についての標本を抜き出せばよい。

# 【母比率推定の標本数】
母比率の区間推定における標本の大きさを決定する。

誤差の比率に注目し、標本の数を決定する。

母比率推定の標本数

(1)標本比率が予測できる時 
$$標本数=(標本比率)×(1-標本比率)× \begin{pmatrix} \frac{k}{誤差の比率} \end{pmatrix}^2$$


(2)標本比率が予測できない時
$$標本数=\frac{1}{4}× \begin{pmatrix} \frac{k}{誤差の比率} \end{pmatrix}^2$$
$$\begin{equation}k=\left \{\begin{array}{l}  90\%=1.65 \\ 95\%=1.96 \\ 99\%=2.58　\end{array}\right.\end{equation}$$


(例１)内閣支持率34%と予想、この支持率を信頼区間95%で推定した場合、誤差を2%以内に抑えるためには、標本数がいくつ必要か？
$$標本数=0.34 × (1-0.34) × \begin{pmatrix} \frac{1.96}{0.02} \end{pmatrix}^2 = 2155$$

(例２)ひまわりの種子の発芽する確率。標本抽出により区間推計したい。信頼度90%で誤差を5%以内におさえるには、何粒実験すればよいか？
誤差比率=0.05 k=90%=1.65 であるので
$$標本数=\frac{1}{4} × \begin{pmatrix} \frac{1.65}{0.05} \end{pmatrix}^2 = 272.55$$
つまり、約272粒について発芽実験をすればよい。

## 【成功確率の推定】

ベルヌーイ母集団から無作為標本を得るとし、$P(X_i-1)=p$, $P(X_i=0)=1-p$ とする。ベルヌーイ確率変数の性質により、母平均はp, 母分散はp(1-p)になる。
成功回数は1が出る回数なので、$\sum_{i=1}^{n}X_i$ 
成功確率ｐの推定量として、平均成功率
$$\hat{p}=\frac{\sum_{i=1}^{n}X_i}{n}$$
を使うが、これは標本平均である。だから、標本平均の性質により
$$E(\hat{p})=p,　V(\hat{p})=\frac{p(1-p)}{n}$$

$\hat(p)$ を使って、pの区間推定を行う。中心極限定理により観測個数が十分大きい場合
$$\frac{\hat{p} - p}{\sqrt{p(1-p)/n}}$$
の分布を、$$N(0, 1)$$ で近似可能。そこで、標準正規近似を用いて信頼区間を作成する。標準正規分布の右確率 $α/2 = z_{α/2}$
分母にふくまれるpを$\hat{p}$で代用すると、信頼係数$100(1-α)\%$の信頼区間は

$$
\begin{eqnarray}
\left[
\begin{array}{ccc}
\hat{p} - z_{α/2} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}},　\hat{p} + z_{α/2}  \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}
\end{array}
\right]
\end{eqnarray}$$

標本分散 $S^2$ を使って、母分散を推定。$X_i$ は 0 と 1 しかとらないので $X_i^2 = X_i$

$$
\begin{eqnarray}
S^2 &=& \frac{\sum_{i=1}^{n}(X_i-\bar{X})^2}{n - 1} \\
    &=& \frac{\sum_{i=1}^{n}X_i^2 - n\bar{X}^2}{n-1} \\
    &=& \frac{\sum_{i=1}^{n}X_i ^ n \hat{p}}{n-1} \\
    &=& \frac{n}{n-1}(\hat{p} - \hat{p}^2)
\end{eqnarray}
$$


したがって、係数を除いて、$S^2$ は $\hat{p}(1-\hat{p})$
ｎが大であるなら、　$S^2$　と　$\hat{p}(1-\hat{p})$ の差は小さい。

(例１)　4000世帯のうち300世帯が番組視聴、この番組の視聴率の区間推定
世帯がその番組をみているか否かは、ベルヌーイ確率分布に従うとする。ｐの推定値は、観測個数4000、視聴世帯数300を用いて
$$\hat{p}=\frac{300}{4000} = 0.075$$

信頼係数95%、標準正規分布の右確率2.5%　$z_{0.025} = 1.96$ 信頼区間の下限上限を求める
$$0.075\pm 1.96 \sqrt{\frac{0.075(1-0.075)}{4000}} = [0.0668, 0.0832]$$
信頼係数95%の信頼区間は、$[6.68\%, 8.32\%]$

### (Y)E(・) ,　V(・)

#### (a)分散と標準偏差(V)

平均は分布の中心位置を示す代表値だが、中心位置の周辺での確率変数の広がり具合を示す代表値として分散がある
散らばりを測りたいから、平均$\mu$から図ったはずれの2乗、$(X-\mu)^2$の加重平均を計算して分散と呼ぶ。

**加重平均に用いる重みは、各座標値における密度や確率が使われる**

**離散確率変数については$(x_1-\mu)^2, (x_2-\mu)^2, ・・・, (x_n-\mu)^2$の平均であるから**

$$
\begin{eqnarray}
\sigma^2 &=& V(E) \\
         &=& E(X-\mu)^2 \\
         &=& (x_1-\mu)^2 p(x_1) + ・・・+(x_m-\mu)^2 p(x_m) \\
         &=& \sum_{i=1}^{n}(x_i-\mu)^2 p(x_i) 
\end{eqnarray}
$$

**連続確率変数については、ｘを積分変換し**

$$
\begin{eqnarray}
\sigma^2 &=& V(X) \\
         &=& E((X-\mu)^2) \\
         &=& \int_{-\infty}^{\infty}(x-\mu)^2 f(x)dx
\end{eqnarray}
$$

と定義。

**確率変数Xの分散はV(X)or Var(X)のように記される**

離散型では分散の意味も説明しやすい。平均から離れた値では、はずれの2乗は大きいが、もしその値を取る確率が非常に小さいならば、その頃には分散にあまり影響しない。
確率関数が平均の周囲の座標で高い値になり、平均から遠く外れた座標で非常に小さい値になるならば、確率関数が正である範囲が広くとも分散は小さくなる。

連続型では、積分が発散する事もある。


**分散の正の平方根は標準偏差(standard deviation)と呼ばれ、よく利用される代表値**

**標準偏差はsd,s.d.,D,σと記される。**


### (b)期待値計算(E)

**離散確率変数では、確率関数を重みとした加重平均**

**連続確率変数では、密度関数を重みとした加重平均**

このような平均や分散の一般的な操作方法が**期待値(expectation)**

**期待値Eは確率変数の関数の平均**

確率変数の任意の関数を$g(X)$と書くなら、離散確率変数では
$$E(g(X)) = g(x_1)p(x_1)+ ・・・+g(x_m)p(x_m)$$

連続確率変数ではxを積分変数とし
$$E(g(X)) = \int_{-\infty}^{\infty}g(x)f(x)dx$$

平均は$g(X)=X$, 分散$g(X)=(X-\mu)^2$と関数ｇを定めた場合の期待値

期待値の操作上便利な性質。$E(X) = \mu$

(1.1)定数cについてE(c)=c.

密度関数の総面積は１なので
$$
\begin{eqnarray}
E(c) &=& \int_{-\infty}^{\infty}cf(x)dx \\
    &=& c\int_{-\infty}^{\infty}f(x)dx \\
    &=& c
\end{eqnarray}
$$


(1.2)$E(X-\mu)=0.$
$$E(X-\mu) = E(X) - E(\mu)$$
上記のように分解可能、第一項は$\mu$, $\mu$が定数であるので、第二項は$\mu$


(1.3)$E(aX+b) = a\mu +b.$
aとbを定数とし、g(X)=aX+bとすれば
$$
\begin{eqnarray}
E(aX+b) &=& \int_{-\infty}^{\infty}(ax+b)f(x)dx \\
        &=& a\int_{-\infty}^{\infty}xf(x)dx + b\int_{-\infty}^{\infty}f(x)dx \\
        &=& a\mu + b
\end{eqnarray}
$$



(1.4)2関数g(X)とh(X)について$E(cg(X) + dh(X)) = cE(g(X)) + dE(h(X))$
c,dは定数

$$
\begin{eqnarray}
E(cg(X) + dh(X)) &=& \int0_{-\infty}^{\infty}(cg(x) + dh(x))f(x)dx \\
                 &=& c\int_{-\infty}^{\infty}g(x)f(x)dx + d\int_{-\infty}^{\infty}h(x)f(x)dx \\
\end{eqnarray}
$$

### (b)期待値　(母分布に依存しない性質)
平均が$\mu$, 分散$\sigma^2$である母集団からのランダムサンプリング。$E(S^2)=\sigma^2$となる

添え字iにかかわりない因子を和の外に出す事により
$$\sum_{i=n}^{n}(X_i - \bar{X})(\bar{X} - \mu) = (\bar{X} - \mu)\sum_{i=1}^{n}(X_i-\bar{X}) = 0$$
となるので
$$
\begin{eqnarray}
\sum_{i=1}^{n}(X_i - \mu)^2 &=& \sum_{i=1}^{n}((X_i - \bar{X} + (\bar{X} - \mu)))^2 \\
                            &=& \sum_{i=1}^{n}(X_i - \bar{X})^2 + n(\bar{X} - \mu)^2 + 2\sum_{i=1}^{n}(X_i-\bar{X})(\bar{X}-\mu)
                            &=& \sum_{i=1}^{n}(X_i - \bar{X})^2 + n(\bar{X} - \mu)^2
\end{eqnarray}
$$

と整理可能。左辺の期待値
$$E\sum_{i=1}^{n}(X_i - \mu)^2 = \sum_{i=1}^{n}E((X_i - \mu)^2) = \sum_{i=1}^{n}V(X_i)=n\sigma^2$$
となり、第二項の期待値は、標本分散のｎ倍であるので、$\sigma^2$になる。整理して、式の第一項の期待値が
$$
E\begin{bmatrix} \sum_{i=1}^{n}(X_i-\bar{X})^2 \end{bmatrix} = (n - 1)\sigma^2
$$

となるので、両辺を(n-1)で割って定理が導き出される。

# 【点推定の規範】

母数を観測値に依存せず、癖なく、無駄なく推定することが望ましい推定量。

観測値をもとにあてずっぽうも考えられるが、これはよい推定量とは言えない。

推定量は確率変数であり、確率変数は分布を持つから、分布の性質として、癖なく、無駄なく推定できなければならない。

大きさｎの無作為標本　$X_1..X_n$　母数　$θ$　の推定量を
$$T(X_1, X_2, X_3 ... X_n) \equiv T(X_n)$$

$\hat{θ}$ と書く場合もある。推定したい母数は、一般的に $θ$ 

---
### 不偏性
推定量$T(X_n)$は確率変数の関数であるから、推定量も確率変数である。したがって推定量の期待値を考えることが可能。




推定量の期待値Eが母数 $θ$ に一致することが、推定量の規範の１つになる
$$E[T(X_n)] = θ$$
がこの規範

この時、推定量$T(X_n)$は、不偏であるという。不偏性をもつ推定量を、不偏推定量

不偏性は癖のなさを意味する。$θ$の推定量の分布が左右対称であるなら、分布の中心が母数$θ$に一致することが不偏性

(例)大きさｎの無作為標本{$X_1, X_2, X_3, ... X_n$}, 母平均 $μ$、したがって $E(X_i)=μ$。 次の推定量が不偏であることを確かめる
$$
\begin{eqnarray}
T_1(X_n) &=& \frac{1}{n}\sum_{i=1}^{n}X_i \\
T_2(X_n) &=& \frac{1}{n+1}\begin{pmatrix} X_1 + \sum_{i=1}^{n}X_i \end{pmatrix}
\end{eqnarray}
$$


$T_1$ の不偏性
$$E[T_1(X_n)] = \frac{1}{n}\sum_{i=1}^{n}E(X_i) = \frac{1}{n}\sum_{i=1}^{n}μ=μ.$$

$T_2$ の不偏性
$$E[T_2(X_n)] = \frac{1}{n+1}\begin{pmatrix}E(X_1) + \sum_{i=1}^{n}E(X_i)\end{pmatrix} = \frac{1}{n+1}\begin{pmatrix}μ+\sum_{i=1}^{n}μ\end{pmatrix}=μ.$$

--------
### 一致性
一致性とは、θの推定量 $T(X_n)$ が観測個数nが大きくなるに連れ、θに近づく性質である。

大数の法則、母平均μと、母分散 $σ^2$ が有限であれば、無作為標本の標本平均$\hat{X}$ は、任意の正の定数 $\varepsilon$ に関して
$$\lim_{n \to ∞}　P[|\hat{X} - μ| \ge \varepsilon] = 0$$

$\hat{X}$ は μ に確率収束するという。

正の値 $\varepsilon$ を固定し、$\varepsilon = kσ/\sqrt{n}$ で、$k = \sqrt{n}\varepsilon/σ$ なので、
$$P[|\bar{X} - μ|\geq \varepsilon]\le\frac{σ^2}{n\varepsilon^2}$$
と変形でき、確率は正または0なので、　$n \to ∞$　の極限を取れば定理が証明可能。

**この定理は、標本平均が母平均の周囲に集中することを示している。これを確立収束と呼び**
$$\DeclareMathOperator*{\plim}{plim}$$
$$\plim_{n \to \infty}\bar{X} = μ$$

確率収束 $\plim_{n \to \infty}$ を使うと、
$$\plim_{n \to \infty}T(X_n) = θ$$

と表記できる。$T(X_n)$ を一致推定量という。

観測個数が無限大に増えるということは、標本が母集団に一致することに等しい。

母平均 $μ$, 母分散 $σ^2$ としたとき。$\bar{X} = \sum_{i=1}^{n} X_i/n$
とすると、上記により、　$\bar{X}$ は μ に確率収束する。

$$\plim_{n \to \infty} \bar{X} = μ$$

$\bar{X}$は一致推定量

$σ^2$の不偏推定量 $S^2$ も一致推定量

$$\plim_{n \to \infty} S^2 = σ^2$$

ｎで割っても一致推定量であることに変わりない。

成功確率推定量で定義された　$\hat{p}$ は、標本平均にほかならないので、pの不偏推定量になる

$$\plim_{n \to \infty} \hat{p} = p$$

が成り立ち、一致性を持つ。

## 効率性

不偏推定量が複数個あれば、分散が小さい推定量を選ぶ。

**分散が小さいということは、母数θに近い値を推定する確率が高い事を意味する。**

したがって推定量　$\hat{θ}$　は　$θ$ に近い可能性が高い。

これが効率性(efficiency) の規範。つまり無駄が少ない。

(例１)：大きさnの無作為標本を$(X_1, X_2, ... X_n)$　とすると、母平均はμ、母分散は$σ^2$
したがって、$E(X_i) = μ,V(X_i)=σ^2$. 上記$T_1, T_2$の分散を計算し、効率性を示す。

$T_1$の分散は、通常の標本平均の分散であり、
$$V[T_1(X_n)] = \frac{1}{n^2}\sum_{i=1}^{n}V(X_i) = \frac{σ^2}{n}$$


$T_2$ には$X_1$ が2個はいっており、また $X_1$ とほかの確率変数は独立に分布するので

$$
\begin{eqnarray}
V[T_2(X_n)] &=& \frac{1}{(n+1)^2}\begin{pmatrix} V(2X_1)+\sum_{i=2}{n}V(X_i) \end{pmatrix} \\
            &=& \frac{1}{(n+1)^2}\begin{pmatrix} 4σ^2 + \sum_{i=2}{n}σ^2 \end{pmatrix} = \frac{n+3}{(n+1)^2}σ^2
\end{eqnarray}
$$

なので、
$$V[T_1(X_n)] \le V[T_2(X_n)]$$

が成り立つ。記号は、n=1のとき成り立ち、$T_1$ のほうが $T_2$ よりも効率的

$T_2$　に余分な項が加わっているので、不偏性が保たれても、効率性が落ちる。こういった無駄をなくすことが望ましいというのが、効率性の規範

**高度だが、諸条件のもとで、不偏推定量の分散の下限が、フィッシャー情報量の逆数によって与えられる。証明には、クラメル=ラオの不等式が使用される**

# 平均２乗誤差(MSE)

$θ$ の推定量　$T(X_n)$, あるいは $\hat{θ}$ が不偏性を持たないとき、推定量の期待値と　$θ$　の差を　**バイアス(偏り)**　という。

$$バイアス = E(\hat{θ} - θ)$$

**バイアスは小さい方がよい。不偏推定量のバイアスは0**

**バイアスと効率性を統合する概念として、平均二乗誤差(mean squared error : MSE)**

$$MSE(\hat{θ}) = E[(\hat{θ}) - θ]$$

MSE
$$
\begin{eqnarray}
MSE(\hat{θ}) &=& E[(\hat{θ} - θ)^2] \\
             &=& E[(\hat{θ} - E(\hat{θ}) + E(\hat{θ}) - θ)^2] \\
             &=& V(\hat{θ}) + [E(\hat{θ}) - θ]^2
\end{eqnarray}
$$


この分解では、$E(\hat{θ}) - θ$ が確率変数でないことに注意。

**第一項は、$\hat{θ}$　の分散であり、第二項はバイアスの二乗. バイアスのない不偏推定量であれば、第二項は0になり、MSEは分散に一致する。**

MSEが大きい推定量は、分散かバイアスのどちらかが大きい推定量となるので望ましいとは言えない。

**バイアスバリアンストレードオフであるので、分散が上がればバイアスは下がるが、バイアスが下がれば分散は上がる。**

MSEは癖と無駄を合算する基準。


(例１)　:　$N(μ, σ^2)$ から求めた無作為標本 $(X_1, X_2, X_3, ..., X_n)$ をもとに、$σ^2$ の推定量$S^2$
$$\hat{σ}^2 = \frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})^2$$
の平均に乗誤差を比較する。

自由度がｋのカイ2乗確率変数の期待値はk、分散は2k。

自由度(n-1)のカイ二乗確率変数で、Uの平均は(n-1)、分散は2(n-1)、$S^2$の期待値は$σ^2$　$S^2$の分散を求める
$$
\begin{eqnarray}
V(S^2) &=& E\begin{bmatrix} \begin{Bmatrix} \frac{1}{1-n}\sum_{i=1}^{n}(X_i-\bar{X})^2 - σ^2\end{Bmatrix}^2 \end{bmatrix} \\
       &=& \begin{pmatrix} \frac{σ^2}{n-1} \end{pmatrix} E\begin{bmatrix} \begin{Bmatrix} U - (n-1) \end{Bmatrix}^2 \end{bmatrix} \\
       &=& 2\frac{σ^4}{n-1}
\end{eqnarray}
$$


なので、$S^2$の期待値でもある。$\hat{σ}^2$の期待値は
$$E(\hat{σ}^2) = \frac{n-1}{n}E(\hat{S}^2) = \frac{n-1}{n}σ^2$$

だから、バイアスは $(-1/n)σ^2$ となる。$\hat{σ}^2$の分散は
$$
\begin{eqnarray}
E\begin{pmatrix} \frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})^2 - \frac{n-1}{n}σ^2 \end{pmatrix} &=& \begin{pmatrix} \frac{σ^2}{n} \end{pmatrix}^2 V(U) \\
  &=& 2(n-1)\begin{pmatrix} \frac{σ^2}{n} \end{pmatrix}^2
\end{eqnarray}
$$


MSEは、分散+バイアスの二乗で

$$2(n-1)\begin{pmatrix} \frac{σ^2}{n} \end{pmatrix}^2 + \begin{pmatrix} \frac{σ^2}{n} \end{pmatrix}^2 = \frac{2n-1}{n^2}σ^4$$

比べると、nが３以上であれば $S^2$ のMSEの方が大きい。あるいはバイアスがある推定量 $\hat{σ}^2$ のほうが、MSEが小さくなる。

# モーメント法
推定法。確率変数の累乗の期待値(積率、モーメント)を確立変数の累乗の標本平均(標本モーメント)で推定、推定量を得る。

母集団の1次と2次のモーメントが、
$$E(X_i) = μ_1$$
$$E(X_i^2) = μ_2$$

であるとする。**ただし、母分布を正規分布のように特定する必要はない**

母集団からランダムサンプリングを得ると仮定し、1次、2次の期待値(モーメント)を、確率変数の累乗の標本平均(標本モーメント)
$$\hat{μ}_1 = \frac{1}{n}\sum_{i=1}^{n}X_i =\bar{X},　\hat{μ}^2 = \frac{1}{n}\sum_{i=1}^{n}X_i^2$$

で推定する。分散$V(X_i)$は
$$V(X_i) = E(X_i^2) - E(X_i)^2 = μ_2 - μ_1^2$$

分散モーメント法推定量は
$$
\begin{eqnarray}
\hat{V}(X_i) = \hat{μ}_2 - \hat{μ}_1^2 &=& \frac{1}{n}\sum_{i=1}^{n}X_i^2 - \bar{X}^2 \\
                                       &=& \frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})^2
\end{eqnarray}
$$


$\hat{σ}^2$ に一致するが、この推定量は不偏ではない。一致性は持つ。

平均周りのモーメントを使ってもよい。
$$
\begin{eqnarray}
E(X_i) &=& μ \\
E[(X_i-μ)^2] &=& m_2 \\
E[(X_i-μ)^3] &=& m_3 \\
E[(X_i-μ)^4] &=& m_4
\end{eqnarray}
$$

を、平均周りの標本モーメント
$$
\begin{eqnarray}
\hat{μ} &=& \frac{1}{n}\sum_{i=1}^{n}X_i - \bar{X} \\
\hat{m}_2 &=& \frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})^2 \\
\hat{m}_3 &=& \frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})^3 \\
\hat{m}_4 &=& \frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})^4
\end{eqnarray}
$$



で推定。平均周りのモーメントを使うと以下、$歪度 = \frac{m_3}{m_2^{\frac{3}{2}}}$ $尖度 = \frac{m_4}{m_2^2}$
$$\frac{\hat{m}_3}{\hat{m}_2^\frac{3}{2}}, \frac{\hat{m}_4}{\hat{m}_2^2}$$

n, n-1　は、モーメント法では分母はｎを使用するが、ｎが大きければ両者の違いは無視可能。

**モーメント法による推定量は、一般に一致制を持つが、不偏性はないことが多い、しかし、母集団の分布型を仮定する必要はないので、推定に当たっては使用しやすい**

**軽量経済学では、モーメント法の一種であるGMM(一般化モーメント法)が広く使用されている**

# 最尤推定法
母数推定において、重要な推定法である最尤推定法(maximum likelihood estimation)

複数の離散確率変数に関しては、同時確率関数、連続確率変数に関しては同時確率密度関数が定義可能。

同時確率関数および同時密度関数とは、母数を所与として、ある観測値が生じる同時確率(密度)を表すもの。

**逆にこれらの関数を、観測された値を所与とし、母数を未知数と考えれば、尤度関数になる。**

**この関数の変数は母数であり、同時確率(密度)が高いほど、母数の値が尤もらしいと理解する。**

このような直観にもとづいて、この尤度(likelihood)を**最大化**する母数値を推定値とするのが、**最尤推定法**

以下ベルヌーイ試行の確率関数を使用し、成功確率を最尤推定法で推定する

### (1)成功確率の最尤推定

ランダムサンプリングの母分布は、成功確率pで１、失敗で０を取るベルヌーイ試行とする

観測値が($x_1, x_2, ...x_n$)であったとする。同時確率はpを所与として
$$
\begin{eqnarray}
P(X_1 = x_1, X_2 = x_2, ... X_n &=& x_n|p) \\
  &=& p(x_1, x_2, ... x_n|p) = \prod_{i=1}^{n}p^{x_i}(1-p)^{1-x_i}
\end{eqnarray}
$$


簡略化、成功回数を $z = \sum_{i=1}^{n}x_i$ と表記し
$$p(x_1, x_2, ... x_n|p) = p^z(1-p)^{n-2}$$

と書ける。成功確率ｐを推定するための尤度関数(likelihood function)
は、上記の同時確率式に、$x_1, x_2, ...x_n$　を所与として、ｐの関数と理解し、
$$L(p|x_1, x_2, ...x_n) = p^z(1-p)^{n-\mathit{z}}$$
となる。

**尤度関数を最大にする母数ｐの値を、最尤推定値とする**

**最尤推定値を簡単に求めるために、尤度の自然対数を取り、対数尤度関数**

$$log \mathit{L}(p) = \mathit{z} log p + (n - \mathit{z})log(1-p)$$

尤度の最大化と、対数尤度関数の最大化は同じ結果をもたらすので、上記最大化するPを最尤推定値とする。
最大化の1次条件は上記の式のｐに関する微分が0になる事


(1次条件)

=============================================================================================
(1.1)残差平方和

残差とは、仮定したｙ、ｘの関係$y=\alpha+\beta x$と観測点($x_i,y_i$)の差。この差は理論と現実の乖離
$\alpha = a, \beta = b$とし、これを所与とする。a,bは推定値。

ある説明変数の値$x_i$に対する直線状のy値を、回帰値$\hat{y}_i$とし、式
$$\hat{y}_i = a + bx_i$$
と定義。ここで観測点($x_i,y_i$)と直線状の点($x_i,\hat{y}_i$)の距離
$$e_i=y_i-\hat{y}_i=y_i-(\alpha+\beta x_i)$$
を残差と呼ぶ。残差を二乗して各観測点に関して和を取った量
$$(残差平方和)RSS = e_1^2 + e_2^2,...e_n^2 = \sum_{i=1}^{n}e_i^2=\sum_{i=1}^{n}(y_i-\alpha-\beta x_i)^2$$
これをRSS(residual sum of squares)、残差平方和

正規方程式を$\alpha$ $\beta$に関するRSS(残差平方和)の最小化(最大化)によって導くとする。

**最小化(最大化)の必要条件としてRSSの$\alpha$ $\beta$ に関する偏微分を0と置く。これを1次条件と呼ぶ**
一般の回帰における正規方程式の解も同じ手続きにより導き出せる。

**偏微分とは、変数$\alpha$ $\beta$の２つがあるが、他方を固定し、一方の微分を求める事。**

$\alpha$に関する1次条件は
$$\frac{\partial RSS}{\partial a} = 0$$

となる。上記は、$\beta$ を固定して、$\alpha$に関する微分を求めると理解してよい。

RSSはn項から成立するので、i番目の項のみを取り上げると、$\beta$を固定するので
$$\frac{\partial(y_i-\alpha-\beta x_i)^2}{\partial \alpha} = (-2)(y_i - \alpha - \beta x_i)$$

式を整理
$$\sum_{i=1}^{n}y_i = na+b\sum_{i=1}^{n}x_i$$
これは最小2乗推定値であり
$$\sum_{i=1}^{n}x_iy_i=a\sum_{i=1}^{n}x_i + b\sum_{i=1}^{n}x_i^2　(1)$$
を満たすこの2式は正規方程式なので、2元連立方程式を満たすaとbが最小2乗推定値となる。

bに関する1次の条件は、
$$\frac{\partial RSS}{\partial b} = 0$$

i番目の項だけを取り上げると、aを固定して
$$\frac{\partial(Y_i - \alpha - bx_i)^2}{\partial b} = (-2x_i)(y_i - \alpha - bx_i)$$

ｂに関する微分になっている。このような項がn個存在するはずなので、和を取ると
$$\frac{\partial RSS}{\partial b} = (-2)\sum_{i=1}^{n}(y_i-\alpha - bx_i)x_i = 0$$

この式を整理すれば上記(１)が導かれる

1次条件を計算する
$$\frac{\partial log \mathit{L}}{\partial p} = \frac{\mathit{z}}{p} = \frac{(n-\mathit{z})}{(1-p)} = 0$$

この式から未知数ｐを求め、$x_i$を$X_i$　($i = 1, 2, ...n$)に置き換えると、最尤推定値は
$$\hat{p}_{ML} = \frac{1}{n}\sum_{i=1}^{n}X_i$$

この推定量は、標本平均に一致

# 尤度関数

最尤推定法の一般的説明

同時確率(密度)関数は、母数を所与として、観測値が生じる確率(密度)を表す。**逆に、この関数を、観測値を所与として母数の関数であると考えれば、尤度関数になる**

**最尤推定法とは観測地を所与として、尤度関数を最大化する*母数*を求める方法**

実際には、計算が簡単になるために、尤度の自然対数を取った対数尤度をもとに、最大化の計算をする。

連続確率変数の場合、ランダムサンプリングの母密度関数が　$f(x|θ)$　であるとする。同時確率密度関数は、密度関数　$f(x_i|θ)$ の積
$$f(x_1, x_2, x_3, ... x_n|θ) = \prod_{i=1}^{n}f(x_i|θ)$$

尤度関数は、上記の同時確率密度関数を、観測地($x1, x2, ...,x_n$)を所与とし、母数θの関数であると考えるから
$$\mathit{L}(θ|x_1, x_2, ..., x_n) = \prod_{i=1}^{n}f(x_i|θ)$$

上記尤度関数の自然対数を取ると、対数尤度関数
$$log \mathit{L}(θ;x_1,x_2,...x_n) = \prod_{i=1}^{n}log f(x_i|θ)$$

最尤推定量は、尤度関数式において、観測地($x_1, x_2, ..., x_n$)を確立変数($X_1,X_2,...X_n$)に置き換え、その上で対数尤度を通して、最大化する$\hat{θ}_{ML}$

正規分布の母数の推定問題に対し、具体的に最尤推定法を見る

#### (1)正規分布の平均と分散の最尤推定
(1.1)正規母集団$N(μ, σ^2)$から、大きさｎのランダムサンプリングを得る。平均μ、分散$σ^2$を最尤推定する

$(X_1, X_2, ...,X_n)$ の同時密度関数は、独立性により各確率変数$X_i$の密度関数の積で表され、以下

$$
\begin{eqnarray}
f(x_1, X_2, ..., x_n|μ,σ^2) &=& \prod_{i=1}^{n}f(x_i|μ,σ^2) \\
  &=& \prod_{i=1}^{n} \begin{Bmatrix} \frac{1}{\sqrt{2\piσ^2}}exp\begin{pmatrix} -\frac{(x_i - μ)^2}{2σ^2} \end{pmatrix} \end{Bmatrix}
\end{eqnarray}
$$


尤度関数は、同時密度関数において観測地を所与として、母数の関数としてみたものなので
$$\mathit{L}(μ,σ^2|x_1,x_2,...x_n) = \prod_{i=1}^{n} \begin{Bmatrix} \frac{1}{\sqrt{2\piσ^2}}exp\begin{pmatrix} -\frac{(x_i - μ)^2}{2σ^2} \end{pmatrix} \end{Bmatrix}$$

そこで、対数尤度関数は、
$$
\begin{eqnarray}
log \mathit{L}(μ,σ^2) &=& \sum_{i=1}^{n} \begin{Bmatrix} -\frac{1}{2}log(2\piσ^2) - \frac{(x_1-μ)^2}{2σ^2} \end{Bmatrix} \\
                      &=& -\frac{n}{2}log(2\piσ^2) - \frac{1}{2\piσ^2}\sum_{i=1}^{n}(x_i-μ)^2
\end{eqnarray}
$$


対数尤度関数を最大化する母数μと$σ^2$。最大化の１次条件は、上記対数尤度関数式をμと$σ^2$に関して微分し、0と置いた連立方程式
$$
\begin{eqnarray}
\frac{\partial log \mathit{L}}{\partial μ} &=& \frac{1}{σ^2}\sum_{i=n}^{n}(x_i - μ) = 0 \\
\frac{\partial log \mathit{L}}{\partial \sigma^2} &=& -\frac{n}{2σ^2} + \frac{1}{2σ^4}\sum_{i=1}^{n}(x_i - μ)^2 = 0
\end{eqnarray}
$$


連立方程式を解き、解の($x_1, x_2, ..., x_n$)を確立変数($X_1, X_2, ..., X_n$)に置き換え、最尤推定量
$$
\begin{eqnarray}
\hat{μ}_{ML} &=& \frac{1}{n}\sum_{i=1}^{n}X_i \\
\hat{σ}_{ML}^{2} &=& \frac{1}{n}\sum_{i=1}^{n}(X_i - \hat{μ}_{ML})^2
\end{eqnarray}
$$

$σ^2_{ML}$ は上記MSEの比較を行った$$\hat{σ}^2 = \frac{1}{n}\sum_{i=1}^{n}(X_i - \hat{X})^2$$

# ベイズ法
標本から得る確率関数or尤度関数を $p(x|θ)$ とする

母数θが事前分布 $\pi(θ)$ に従属すると仮定する。上記ベイズ公式により、母数θの事後分布を
$$\pi(\theta|x) = \frac{p(x|\theta)\pi(\theta)}{\int p(x|\theta)\pi(\theta)d\theta}$$

分子が条件付き密度と周辺密度の積で、(x,θ)の同時密度

分母が同時密度をθに関して積分しているのでxの周辺密度になっている。

離散確率変数を考慮した場合分母が和になる。
(５)

ここでは、連続確率変数θを考えるために、積分になっている。

ベイズ法は、分母θの事前分布$\pi(\theta)$を標本の尤度を利用し、事後分布$\pi(\theta|x)$に更新する。

最後に事後分布を用いて推論を行う。θのベイズ推定量は、事後分布に関するθの期待値
$$\hat{\theta} = E(\theta) = \int \theta\pi(\theta|x)d\theta$$

これを事後分布という。

#### 成功確率のベイズ推定

ベルヌーイ試行の成功確率ｐのベイズ推定

ベルヌーイ試行をn回試行した場合の確率関数は、成功する回数を $\mathit{z}=\sum_{i=1}^{n}x_i$と置くと
$$p(x_1, x_2, ..., x_m|p) = \prod_{i=1}^{n}p^{x_i}(1-p)^{1-x_i} = p^{\mathit{z}}(1-p)^{n-\mathit{z}},　0\le p\le1$$

これは上記でも記したように尤度関数でもある。

ベルヌーイ分布に対応する事後分布として、ベータ分布
$$\pi(p) = \frac{1}{B(\alpha,\beta)}p^{\alpha-1}(1-p)^{\beta-1}$$

を使う。これは主観をもとに分析者が選ぶ事前分布が、ベータ分布であるとしている。$ベータ関数 = B(\alpha, \beta)$
$$B(\alpha, \beta) = \int_{0}^{1}x_{\alpha-1}(1-x)^{\beta-1}dx,　\alpha,\beta > 0$$

ガンマ関数によって
$$B(\alpha, \beta) = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha+\beta)}$$

ガンマ関数の$\Gamma(\alpha)$
$$\Gamma(\alpha) = \int_{0}^{\infty}x^{\alpha-1}e^{-x}dx,　\alpha > 0$$

ベータ分布から得る事前平均
$$E(p) = \int_{0}^{1}p\pi(p)dp = \frac{\alpha}{\alpha+\beta}$$

**事前分布にベータ分布を使用すると、事後分布もまたベータ分布になる。**
**このように事前事後の分布が同じ型になる分布を、自然共役分布と呼ぶ**

ベイズ法では、計算の便利さのような技術的問題から自然共役分布を使う事が多い。(注)主観によって選ぶという原則からはズレる

上記の母数θの事後分布公式を使用して、事後分布の計算を行うと以下
$$\pi(p|\mathit(z)) = \frac{1}{B(\alpha + \mathit(z), \beta+n-\mathit{z})}p^{\alpha+\mathit{z}-1}(1-p)^{\beta+n-\mathit{z}-1}$$

事後平均によりθのベイズ推定量を求めると、$\mathit{w} = (\alpha + \beta)/(\alpha + \beta + n)$で
$$
\begin{eqnarray}
\tilde{p} = E(p) &=& \int_{0}^{1}p\pi(p|\mathit{z})dp = \frac{\alpha + \mathit{z}}{\alpha+\beta+n} \\
  &=& \mathit{w}\begin{pmatrix} \frac{\alpha}{\alpha+\beta}+(1 - \mathit{w})\begin{pmatrix} \frac{\mathit{z}}{n} \end{pmatrix} \end{pmatrix}
\end{eqnarray}
$$

ベイズ推定量$\tilde{p}$は、事前平均式と、pの標本平均推定量 $\mathit{z}/n$ の加重平均になっている。
この結果でもベイズ法は、事前分布の情報に、標本の情報を加えて推定している事が確認できる。
